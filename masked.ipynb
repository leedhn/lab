```
# p2u 파일 다운로드 
# 페르소나들을 set 형태로 변환 =>persona_set
import json
import collections
with open('/content/drive/MyDrive/Dataset/p2u_pu_for_comet_6.json') as json_file:
    json_data=json.load(json_file)
    value_list={}
    #persona_list=list(json_data.keys())
    persona_set=set(json_data.keys())
    #for key in json_data.keys():
    #    value_list.update(dict.fromkeys(json_data[key],key))
    #print(persona_list)
    #for p in persona_set:
        #print(p)
    #print(persona_set)
    if 'i like to go hunting.' in persona_set:
        print('yes')
```

```
#common_train, common_valid, common_test 는 persona_set과 데이터파일에 공통으로 들어있는 페르소나에 대해서,
#{key:value}=[i]번째 대화:대화의 공통 페르소나} 딕셔너리를 가진다. 
#dt_persona는 train, valid, test 전체에서 persona_set과 공통된 페르소나를 가지도록 한다.
common_train={}
dt_persona=[]
for i in range(len(res_train)):
    r = []
    for p in res_train[i]:
        if p in persona_set:
            dt_persona.append(p)
            r.append(p)
    if len(r)>0:
        common_train[i]=r
```

```
common_valid={}
dt_c_valid=[]
for i in range(len(res_valid)):
    r = []
    for p in res_valid[i]:
        if p in persona_set:
            dt_persona.append(p)
            r.append(p)
            dt_c_valid.append(p)
    if len(r)>0:
        common_valid[i]=r
```

```
common_test={}

for i in range(len(res_test)):
    r = []
    for p in res_test[i]:
        if p in persona_set:
            dt_persona.append(p)
            r.append(p)
    if len(r)>0:
        common_test[i]=r  #res_test의 key와 common_test의 key가 일치함.
```

```
dt_persona=set(dt_persona)
```

```
len(dt_c_valid)
```

```
#masking txt 파일 만들기
#구드라에 파일 복제를 미리 해둔 후 셀 실행했음
# dt_persona(persona_set과 기존 데이터에서 공통으로 가지는 페르소나들)를 기준으로 행을 삭제함
with open('/content/drive/MyDrive/persona/personachat/valid_self_original(1).txt', 'r+') as f: # file을 열고 알아서 닫아 줌 
    lines = [] 
    new_line = '' 
    for line in f: 
        line_s=re.split('your persona: |\n', line)
        if line_s[1] in dt_persona:
            lines=lines+[new_line]
        else: 
            lines = lines + [line] 
    f.seek(0) # file pointer 위치를 처음으로 돌림 
    f.writelines(lines) # 수정한 lines를 파일에 다시 씀 
    f.truncate() # 현재 file pointer 위치까지만 남기고 나머지는 정리
```

```
#answer만 모아서 txt, json파일 만드는 부분
# 페르소나들을 set 형태로 변환 =>persona_set
import json
import collections
import re


with open('/content/drive/MyDrive/Dataset/p2u_pu_for_comet_6.json') as json_file: #p2u 파일 저장된 경로 삽입
    json_data=json.load(json_file)
    value_list={}
    persona_set=set(json_data.keys())


#valid에서 정답지만 뽑아서 저장
ans_valid=[] #정답 문장 끝에 \n 있음
ans_valid_no_enter=[] #정답 문장 끝에 \n 없음
i=0
with open('/content/drive/MyDrive/persona/personachat/valid_self_original.txt') as f: #persona chat 데이터 파일 경로
    for line in f:
        i+=1
        if 'your persona' not in line:
            ans_valid.append(re.split('\\|', line)[-1])
            ans_valid_no_enter.append(re.split('\n|\\|', line)[-2])
    print(i)
same=True
for i in ans_valid:
    if same==False:
        break
    ans = re.split('\n',i)[0]
    #print(ans)
    if ans not in ans_valid_no_enter:
        same=False

if same==True:
    print('ans_valid_no_enter and ans_valid have same answer sentences')
#valid_self_original answer 
#TEXT FILE
with open('/content/drive/MyDrive/persona/personachat/answer_valid_self_original.txt', 'w+') as f: # file을 열고 알아서 닫아 줌 
    f.seek(0) # file pointer 위치를 처음으로 돌림 
    f.writelines(ans_valid) # 수정한 lines를 파일에 다시 씀 
    f.truncate() # 현재 file pointer 위치까지만 남기고 나머지는 정리  

#JSON FILE
with open('/content/drive/MyDrive/persona/personachat/answer_valid_self_original.json', 'w+') as json_file:
    json.dump(ans_valid_no_enter, json_file)
```
